```python
def toffoli_gate():
    Toffoli = QuantumCircuit(3)
    Toffoli.h(2)
    Toffoli.cx(1, 2)
    Toffoli.tdg(2)
    Toffoli.cx(0, 2)
    Toffoli.t(2)
    Toffoli.cx(1, 2)
    Toffoli.tdg(2)
    Toffoli.cx(0, 2)
    Toffoli.t(1)
    Toffoli.t(2)
    Toffoli.cx(0, 1)
    Toffoli.t(0)
    Toffoli.tdg(1)
    Toffoli.h(2)
    Toffoli.cx(0, 1)

    display(Toffoli.draw('mpl'))
    U = Operator(Toffoli)
    display(array_to_latex(U))
```

```python
toffoli_gate()
```

```python
def and_gate():
    qc = QuantumCircuit(3)
    qc.ccx(0, 1, 2)
    return qc
display(and_gate().draw('mpl'))
```

```python
def or_gate():
    qc = QuantumCircuit(3)
    qc.x(range(2))
    qc.ccx(0, 1, 2) # qc.compose(and_gate(), inplace=True)
    qc.x(2)
    return qc
display(or_gate().draw('mpl'))
```

```python
def xor_gate():
    qc = QuantumCircuit(3)
    qc.cx(0, 2)
    qc.cx(1, 2)
    return qc
display(xor_gate().draw('mpl'))
```

```python
def fanout_gate():
    qc = QuantumCircuit(2)
    qc.cx(0, 1)
    return qc
display(fanout_gate().draw('mpl'))
```
